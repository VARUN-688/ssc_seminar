```python
df=pd.read_csv(r'C:\Users\varun\PycharmProjects\ssc_presentation\car_data2.csv')
```

```python
grp_by=df.groupby(by='make',as_index=False)
grp_by
for i,j in grp_by:
    print(i,j.price)
```

```python
grp2=df.groupby(by=['make','body_style'],as_index=False).price.mean()
grp2
```

```python
grp3=df.groupby(by=['make','body_style'],as_index=True).price.mean()
grp3
```

```python
grp4=df.groupby(by=['make','body_style'],as_index=False).price.agg('mean').sort_values(by='price')
grp4
```
